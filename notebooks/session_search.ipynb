```python
import flywheel
fw = flywheel.Client()
project = fw.lookup('davis/presurgicalEpilepsy')

sub_ses_dict={}
for ses in project.sessions():
    session = fw.get(ses.id)
    sub = fw.get(session.parents['subject']).label
    pair='%s: %s \n' % (sub, ses.label)
    for a in session.analyses:
          if a.gear_info.name == 'fmriprep' and "SB_fmriprep" in a.label and a.job.state=='complete':
            print(pair)
            sub_ses_dict.update({sub: ses.label})
```
