```python
import flywheel
fw = flywheel.Client()
proj = fw.lookup('davis/presurgicalEpilepsy')
# create subject list
fin = 'patients.txt'
sid_list = [line.rstrip('\n').split()[0] for line in open(fin)]
for sub in sid_list:
    ses = fw.lookup('davis/presurgicalEpilepsy/%s/ses-01' % (sub))
    for analysis in ses.analyses:
        if "csv" in analysis.label:
            print("%s: %s" % (sub,analysis.job.state))
```
